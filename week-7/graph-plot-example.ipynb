```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# draw a graph between cities and color by PageRank
pairs1 = pd.read_csv('/home/hee3be/Dev/data-science/Untitled Folder/networkx_city_pairs.csv', names=['city1', 'city2'])
pairs2 = pd.read_csv('/home/hee3be/Dev/data-science/Untitled Folder/networkx_city_pairs.csv', names=['city2', 'city1'])
pairs = pd.concat([pairs1, pairs2])
```

```python
# create adjacency matrix
adjacency = pd.crosstab(pairs['city1'], pairs['city2'])
adjacency
```

```python
# create graph object with NetworkX library
g = nx.from_pandas_adjacency(adjacency)
g
```

```python
# run PageRank algorithm and add result to matrix
rank = nx.pagerank(g, alpha=0.85)
adjacency['rank'] = pd.Series(rank)
```

```python
adjacency = adjacency.sort_values(by='rank', ascending=False)
print(adjacency)
```
