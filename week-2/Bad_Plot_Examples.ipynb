# The Plot, the Bad and the Ugly

**This notebook contains examples of VERY BAD plots. Find out what is wrong with each of them.**

```python
import seaborn as sns
from matplotlib import pyplot as plt

sns.set_theme(style="whitegrid")
```

```python
df = sns.load_dataset('penguins')
df.head(3)
```

### Example 1: Species Count

```python
counts = df['species'].value_counts()
```

```python
plt.figure(figsize=(10, 7))
counts.plot.barh(fontsize=16)
plt.xlim(50, 160)
plt.title('Number of penguins by species', fontsize=20)
None

# metric missing
# barchart vertical
```

### Example 2: Bill Sizes

```python
plt.figure(figsize=(10, 7))
sns.boxplot(x="species", y="bill_length_mm", palette=["b"], data=df)
sns.boxplot(x="species", y="bill_depth_mm", palette=['g'], data=df)
plt.xlabel('Species', fontsize=16)
plt.ylabel('bill depth (green) and bill length (blue)', fontsize=16)
plt.title('Bill dimensions of penguins', fontsize=20)
None

# distribution over all 4 quartiles, middle line = medium <=> mean
# better to divide it into 2 plots/ duplicate y axis, bad idea for 90 % of time
```

### Example 3: Comparing two Dimensions

```python
plt.figure(figsize=(10, 7))
sns.lineplot(data=df, x='body_mass_g', y='bill_length_mm', hue='species')
plt.xlabel("Body Mass [g]", fontsize=16)
plt.ylabel("Bill Length [mm]", fontsize=16)
plt.title("Bill length over body mass", fontsize=20)
None

# overlapping standard deviation
# line plot does not make sense if we don't have some kind of time series
```

### Example 4: Histograms

```python
plt.figure(figsize=(10, 7))

df.groupby('species')['body_mass_g'].plot.hist(bins=20, fontsize=16, legend=True)
plt.xlabel("Body Mass [g]", fontsize=16)
plt.ylabel("Frequency", fontsize=16)
plt.title("Distribution of body mass", fontsize=20)
None

# Frequency metric
# overlapping bars
```
