```python
import pandas as pd
import seaborn as sb

entries = pd.read_csv('data/25_10_2022.csv')
entries.columns = ['price', 'ticket type', 'age', '?', 'date', 'departure station', 'destination station', 'duration', 'departure time', 'arrival time']
entries.head()
```

```python
entries.info()
```

```python
entries['price'] = entries['price'].apply(lambda x: x.replace(',', '.'))
entries['arrival time'] = pd.to_datetime(entries['arrival time'],format= '%H:%M' ).dt.time
entries['departure time'] = pd.to_datetime(entries['departure time'],format= '%H:%M' ).dt.time

entries['price'].astype(float)
entries.head()
```

```python
entries.plot(x='departure time', y='price')
```
