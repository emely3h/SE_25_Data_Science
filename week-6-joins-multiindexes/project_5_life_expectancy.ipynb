```python
def pop_convert(x):
    if x.endswith('k'):
        return float(x[:-1])
    else:
        return float(x[:-1]) * 1000
pop = pop.applymap(pop_convert)
pop
```

### Task 7 + 11 + 12

```python
continents = pd.read_csv('data/continents.csv', index_col=1, sep=';')
continents
```

```python
def fert_life_plot(fert_df, life_df, year):
    plt.figure()
    # get fertility and life expectation for year
    life_for_year = life_df[year].to_frame()
    fert_for_year = fert_df[year].to_frame()
    fert_for_year.index.names = ['country']
    # merge fertility and life expectation for year in one df
    merged_df = pd.merge(life_for_year, fert_for_year, left_index=True, right_index=True)
    # add population 
    merged_df = pd.merge(merged_df, pop[year], left_index=True, right_index=True)
    # add continent
    merged_df = pd.merge(merged_df, continents['continent'], left_index=True, right_index=True)
    merged_df.columns = ['life expectancy', 'fertility rate', 'population', 'continent']
    # create plot
    plot = sb.scatterplot(data=merged_df, x='fertility rate', y='life expectancy', hue='continent', size='population',  sizes=(40, 400) )
    #plot = merged_df.plot.scatter(x='fertility rate', y='life expectancy')
    # styling
    plot.set(xlim = (0,10))
    plot.set(ylim=(0,90))
    plot.legend(fontsize=5)
    return plot

test_plot = fert_life_plot(fert, life, 1965)
```

### Task 8

```python
def create_save_plots_in_range(start, end):
    for year in range(start, end):
        plot = fert_life_plot(fert, life, year)
        fig = plot.get_figure()
        fig.savefig(f'plots/{year}.png')
create_save_plots_in_range(1960, 2010)
# when using the seaborn scatter plot in task 7 instead of the pandas scatter plot, something gets messed up in the loop
# while it works for just one test plot when trying to create multiple plots and save them to file 
# each plot also contains the data of the previous plots
```
